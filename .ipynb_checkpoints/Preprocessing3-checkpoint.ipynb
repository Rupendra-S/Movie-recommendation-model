```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = 'f67cf462daa20624a35f1fe8a0798b67'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=f67cf462daa20624a35f1fe8a0798b67'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = ''
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
df['Genre'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
def get_director(x):
    if '(director)' in x:
        return x.split('(director)')[0]
    elif '(directors)' in x:
        return x.split('(directors)')[0]
    else:
        return x.split('(director/screenplay)')[0]
```

```python
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split('screenplay);')[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name'] = df_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split('screenplay);')[-1]).split(",")) < 2:
        return np.NaN
    else:
        return ((x.split('screenplay);')[-1]).split(",")[1])
```

```python
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split('screenplay);')[-1]).split(",")) < 3:
        return np.NaN
    else:
        return ((x.split('screenplay);')[-1]).split(",")[2])
```

```python
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_movies = df_till_2019.drop(columns=['genres','director_name','actor_1_name','actor_2_name','actor_3_name'])
df_movies
```

```python
df_movies.comb=df_movies.comb.astype(str)
```

```python
df_movies['comb'] = df_movies['comb'].apply(lambda x:x.lower())
```

```python
df_movies
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df_movies['comb'] = df_movies['comb'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(df_movies['comb']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = df_movies[df_movies['movie_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df_movies.iloc[i[0]].movie_title)
```

```python
recommend('mom')
```
